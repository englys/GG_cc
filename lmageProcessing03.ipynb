```python
import numpy as np
import matplotlib.pyplot as plt
import rembg
from rembg import remove
from PIL import Image
```

กำหนด path ของ input

```python
path_input = '/content/gdrive/MyDrive/ImageProcessing_Feb2024/images/input/'
```

กำหนด path ของ output

```python
path_output = '/content/gdrive/MyDrive/ImageProcessing_Feb2024/images/output/'
```

```python
# Load รูป จาก Drive
# *** Path ที่เก็บรูป gfg ใน Google Drive ของแต่ละคนไม่เหมือนกัน ***
path_input_gfg = path_input + 'gfgimage.png'
gfg_image = Image.open(path_input_gfg)
plt.axis("off")
plt.imshow(gfg_image)
```

Remove background

วิธีที่ 1

```python
# แปลงรูปเป็น numpy array
gfg_array = np.array(gfg_image)
# Remove background ด้วย rembg
output_gfgarray = rembg.remove(gfg_array)
# สร้าง PIL image จาก output_gfg_array
output_gfgimage = Image.fromarray(output_gfgarray)
# Save output_gfgimage
path_output_gfg = path_output + 'output_gfg.png'
output_gfgimage.save(path_output_gfg)
# Show ภาพผลลัพธ์ที่ได้
plt.axis("off")
plt.imshow(output_gfgimage)
```

```python
wolf_image = Image.open(path_input + 'wolf.jpg')
plt.axis("off")
plt.imshow(wolf_image)
```

```python
path_output_wolf = path_output + 'output_wolf.png'
output_wolf = rembg.remove(wolf_image)
output_wolf.save(path_output_wolf)
plt.axis("off")
plt.imshow(output_wolf)
```
