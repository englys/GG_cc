```python
# Import library / module ที่จะใช้
import numpy as np
import matplotlib.pyplot as plt

import skimage
from skimage import data
from PIL import Image, ImageFilter, ImageEnhance, ImageOps
```

กำหนด Path สำหรับ Input และ Output

```python
path_input = '/content/gdrive/MyDrive/ImageProcessing_Feb2024/images/input/'
path_output = '/content/gdrive/MyDrive/ImageProcessing_Feb2024/images/output/'
```

```python
path_lena = path_input + 'lena.png'
lena = Image.open(path_lena)
plt.axis("off")
plt.imshow(lena)
```

Invert

```python
lena = Image.open(path_lena)
plt.axis("off")
inverted_lena = ImageOps.invert(lena)
inverted_lena.save(path_output + 'inverted_lena.png')
plt.imshow(inverted_lena)
```

Blur

```python
lena = Image.open(path_lena)
plt.axis("off")
blurred_lena = lena.filter(ImageFilter.BLUR)
blurred_lena.save(path_output + 'blurred_lena.png')
plt.imshow(blurred_lena)
```

Grayscale

```python
lena = Image.open(path_lena)
plt.axis("off")
grayed_lena = ImageOps.grayscale(lena)
grayed_lena.save(path_output + 'grayed_lena.png')
plt.imshow(grayed_lena)
```

Contrast

```python
lena = Image.open(path_lena)
plt.axis("off")
contrasted_lena = ImageOps.autocontrast(lena, 20, None)
contrasted_lena.save(path_output + 'contrasted_lena.png')
plt.imshow(contrasted_lena)
```

```python
import cv2
# ต่อภาพแนวตั้ง
img_lena = cv2.imread(path_lena)
im_v = cv2.vconcat([img_lena, img_lena])
# Save รูปที่ต่อ ลงใน output
lena_concat_v = cv2.imwrite(path_output + 'lena_v.png', im_v)
```

```python
import cv2
# ต่อภาพแนวนอน
img_lena = cv2.imread(path_lena)
im_h = cv2.hconcat([img_lena, img_lena])
# Save รูปที่ต่อ ลงใน output
lena_concat_h = cv2.imwrite(path_output + 'lena_h.png', im_h)
```
